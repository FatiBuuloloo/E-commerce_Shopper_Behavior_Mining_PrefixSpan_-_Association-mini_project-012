```python
import pandas as pd
import json
from google.colab import files
```

```python
data1 = pd.read_csv('/content/2019-Oct.csv')
#data2 = pd.read_csv('/content/2019-Nov.csv')
data1.columns
```

```python
data1.shape
```

```python
(data1.isna().sum() / len(data1)) * 100
```

```python
data1["event_time"] = pd.to_datetime(data1["event_time"]).dt.tz_localize(None)
data1.dropna(subset="user_session",inplace=True)
```

```python
print("Number of Unique Users:", data1['user_id'].nunique())
print("Number of Unique Session:", data1['user_session'].nunique())
print("Number of Event:\n", data1['event_type'].value_counts())
```

```python
data1 = data1.sort_values(by=["user_id","event_time"]).reset_index(drop=True)
data1 = data1.iloc[22000004:,:]
```

```python
data1["brand"] = data1["brand"].fillna("Unknown")
data1["event_action"] = data1["event_type"]+"_"+data1["brand"]
```

```python
data1.head(5)
```

```python
sequence_event = data1.groupby(['user_id', 'user_session'])['event_action'].apply(list).tolist()
sequence_event = list(filter(lambda x: len(x)!=1, sequence_event))
```

```python
with open('sequence_event2.json', 'w') as f:
    json.dump(sequence_event, f)
files.download('sequence_event2.json')
```
